```python
x=dum.drop(["Target"],axis=1)
```

```python
y=dum["Target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
```

```python
x_train_standradized=st.fit_transform(x_train)
x_test_standradized=st.transform(x_test)
x_train_standradized=pd.DataFrame(x_train_standradized)
x_test_standradized=pd.DataFrame(x_test_standradized)
```

```python
x_train_standradized
```

```python
colo=["Age","FrequentFlyer","ServicesOpted","AccountSyncedToSocialMedia","BookedHotelOrNot","AnnualIncomeClass_High Income","AnnualIncomeClass_Low Income","AnnualIncomeClass_Middle Income"]
```

```python
x_train_standradized.columns=colo
x_test_standradized.columns=colo
```

```python
x_test_standradized
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
D=DecisionTreeClassifier()
```

```python
D.fit(x_train_standradized,y_train)
```

```python
predict=D.predict(x_test_standradized)
```

```python
predict
```

```python
from sklearn import metrics
```

```python
print("Accuracy :",metrics.accuracy_score(y_test,predict))
```

```python
print(metrics.confusion_matrix(y_test,predict))
```
